```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
no_prune = pd.read_csv('./Model_Metrics/Inference_Metrics/test_history_NO_PRUNE.csv')
prune_high = pd.read_csv('./Model_Metrics/Inference_Metrics/test_history_unstructured_PRUNE_High_Settings.csv')
prune_low = pd.read_csv('./Model_Metrics/Inference_Metrics/test_history_unstructured_PRUNE_Low_Settings.csv')

np_color = 'brown'
np_secondary = 'salmon'

hp_color = 'forestgreen'
hp_secondary = 'lightgreen'

lp_color = 'steelblue'
lp_secondary = 'lightskyblue'
```

```python
plt.figure(figsize=(6, 6))
plt.bar(x='No Pruning', height=no_prune['latency'], width=.5, color=np_color)
plt.bar(x='Light Pruning', height=prune_low['latency'], width=.5, color=lp_color)
plt.bar(x='High Pruning', height=prune_high['latency'], width=.5, color=hp_color)
plt.ylabel('Latency (Seconds)')
plt.xlabel('Pruning Type')
plt.title('Latency (Time for Single Inference) VS Pruning Type')
plt.ylim(.065, .075)
plt.grid(True, axis='y', alpha=.75)
plt.show()

plt.figure(figsize=(6, 6))
plt.bar(x='No Pruning', height=no_prune['throughput'], width=.5, color=np_color)
plt.bar(x='Light Pruning', height=prune_low['throughput'], width=.5, color=lp_color)
plt.bar(x='High Pruning', height=prune_high['throughput'], width=.5, color=hp_color)
plt.ylabel('Throughput (Inferences/Second)')
plt.xlabel('Pruning Type')
plt.title('Throughput VS Pruning Type')
plt.grid(True, axis='y', alpha=.75)
plt.ylim(10, 15)
plt.show()
```

```python
plt.figure(figsize=(6, 6))

# No Pruning 
plt.bar(x='F1', height=no_prune['test_f1'], width=.5, color=np_color, label='No Pruning')
plt.bar(x='Precision', height=no_prune['test_precision'], width=.5, color=np_color)
plt.bar(x='Recall', height=no_prune['test_recall'], width=.5, color=np_color)
plt.bar(x='Accuracy (%)', height=(no_prune['test_acc'] * .01), width=.5, color=np_color)

# Light Pruning
plt.bar(x='F1', height=prune_low['test_f1'], width=.5, color=lp_color, label='Light Pruning')
plt.bar(x='Precision', height=prune_low['test_precision'], width=.5, color=lp_color)
plt.bar(x='Recall', height=prune_low['test_recall'], width=.5, color=lp_color)
plt.bar(x='Accuracy (%)', height=(prune_low['test_acc'] * .01), width=.5, color=lp_color)
plt.ylabel('Score')
plt.xlabel('Score Type')

# High Pruning
# plt.bar(x='F1', height=prune_high['test_f1'], width=.5, color=hp_color, label='High Pruning')
# plt.bar(x='Precision', height=prune_high['test_precision'], width=.5, color=hp_color)
# plt.bar(x='Recall', height=prune_high['test_recall'], width=.5, color=hp_color)
# plt.bar(x='Accuracy (%)', height=(prune_high['test_acc'] * .01), width=.5, color=hp_color)
# plt.ylabel('Score')
# plt.xlabel('Score Type')

# plt.xticks(rotation=50)
# plt.yticks([.2, .6, .8, .85, .9 , .95])
plt.title('F1, Precision, Recall VS Pruning Type')
plt.ylim(.85, .92)
plt.grid(True, axis='y', alpha=.75)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(6, 6))
plt.bar(x='F1', height=prune_high['test_f1'], width=.5, color=hp_color, label='High Pruning')
plt.bar(x='Precision', height=prune_high['test_precision'], width=.5, color=hp_color)
plt.bar(x='Recall', height=prune_high['test_recall'], width=.5, color=hp_color)
plt.bar(x='Accuracy (%)', height=(prune_high['test_acc'] * .01), width=.5, color=hp_color)
plt.ylabel('Score')
plt.xlabel('Score Type')

plt.title('F1, Precision, Recall VS High Pruning')
plt.ylim(.25, .36)
plt.grid(True, axis='y', alpha=.75)
plt.legend()
plt.show()
```
