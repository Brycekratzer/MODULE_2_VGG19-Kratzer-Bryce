```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
np = pd.read_csv('../Model_Metrics/Inference_Metrics/test_history_NO_PRUNE.csv')
lp = pd.read_csv('../Model_Metrics/Inference_Metrics/test_history_unstructured_prune_low_setting.csv')
mp = pd.read_csv('..//Model_Metrics/Inference_Metrics/test_history_unstructured_prune_medium_setting.csv')
hp = pd.read_csv('../Model_Metrics/Inference_Metrics/test_history_unstructured_prune_high_setting.csv')

np_color = 'brown'
np_secondary = 'salmon'

hp_color = 'forestgreen'
hp_secondary = 'lightgreen'

mp_color = 'darkorange'
mp_secondary = 'moccasin'

lp_color = 'steelblue'
lp_secondary = 'lightskyblue'
```

```python
plt.figure(figsize=(6, 6))
plt.bar(x='No Pruning', height=np['latency'], width=.5, color=np_color)
plt.bar(x='Light Pruning', height=lp['latency'], width=.5, color=lp_color)
plt.bar(x='Medium Pruning', height=mp['latency'], width=.5, color=mp_color)
plt.bar(x='High Pruning', height=hp['latency'], width=.5, color=hp_color)
plt.ylabel('Latency (Seconds)')
plt.xlabel('Pruning Type')
plt.title('Latency (Time for Single Inference) VS Pruning Type')
plt.ylim(.065, .075)
plt.grid(True, axis='y', alpha=.75)
plt.show()

plt.figure(figsize=(6, 6))
plt.bar(x='No Pruning', height=np['throughput'], width=.5, color=np_color)
plt.bar(x='Light Pruning', height=lp['throughput'], width=.5, color=lp_color)
plt.bar(x='Medium Pruning', height=mp['throughput'], width=.5, color=mp_color)
plt.bar(x='High Pruning', height=hp['throughput'], width=.5, color=hp_color)
plt.ylabel('Throughput (Inferences/Second)')
plt.xlabel('Pruning Type')
plt.title('Throughput VS Pruning Type')
plt.grid(True, axis='y', alpha=.75)
plt.ylim(10, 15)
plt.show()
```

```python
plt.figure(figsize=(6, 6))

# No Pruning 
plt.bar(x='F1', height=np['test_f1'], width=.5, color=np_color, label='No Pruning')
plt.bar(x='Precision', height=np['test_precision'], width=.5, color=np_color)
plt.bar(x='Recall', height=np['test_recall'], width=.5, color=np_color)
plt.bar(x='Accuracy (%)', height=(np['test_acc'] * .01), width=.5, color=np_color)

# Light Pruning
plt.bar(x='F1', height=lp['test_f1'], width=.5, color=lp_color, label='Light Pruning')
plt.bar(x='Precision', height=lp['test_precision'], width=.5, color=lp_color)
plt.bar(x='Recall', height=lp['test_recall'], width=.5, color=lp_color)
plt.bar(x='Accuracy (%)', height=(lp['test_acc'] * .01), width=.5, color=lp_color)
plt.ylabel('Score')
plt.xlabel('Score Type')

# Medium Pruning
plt.bar(x='F1', height=mp['test_f1'], width=.5, color=mp_color, label='Medium Pruning')
plt.bar(x='Precision', height=mp['test_precision'], width=.5, color=mp_color)
plt.bar(x='Recall', height=mp['test_recall'], width=.5, color=mp_color)
plt.bar(x='Accuracy (%)', height=(mp['test_acc'] * .01), width=.5, color=mp_color)
plt.ylabel('Score')
plt.xlabel('Score Type')

# # High Pruning
# plt.bar(x='F1', height=hp['test_f1'], width=.5, color=hp_color, label='High Pruning')
# plt.bar(x='Precision', height=hp['test_precision'], width=.5, color=hp_color)
# plt.bar(x='Recall', height=hp['test_recall'], width=.5, color=hp_color)
# plt.bar(x='Accuracy (%)', height=(hp['test_acc'] * .01), width=.5, color=hp_color)
# plt.ylabel('Score')
# plt.xlabel('Score Type')

# plt.xticks(rotation=50)
# plt.yticks([.2, .6, .8, .85, .9 , .95])
plt.title('F1, Precision, Recall VS Pruning Type')
plt.ylim(.85, .92)
plt.grid(True, axis='y', alpha=.75)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(6, 6))
plt.bar(x='F1', height=hp['test_f1'], width=.5, color=hp_color, label='High Pruning')
plt.bar(x='Precision', height=hp['test_precision'], width=.5, color=hp_color)
plt.bar(x='Recall', height=hp['test_recall'], width=.5, color=hp_color)
plt.bar(x='Accuracy (%)', height=(hp['test_acc'] * .01), width=.5, color=hp_color)
plt.ylabel('Score')
plt.xlabel('Score Type')

plt.title('F1, Precision, Recall VS High Pruning')
plt.ylim(.7, .75)
plt.grid(True, axis='y', alpha=.75)
plt.legend()
plt.show()
```
